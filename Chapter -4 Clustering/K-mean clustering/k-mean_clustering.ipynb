# K- Means Clustering 

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
# Read the dataset from the CSV file 'Mall_Customers.csv' into a DataFrame
dataset = pd.read_csv('Mall_Customers.csv')

# Select the 4th and 5th columns (index 3 and 4) from the dataset and store them in variable X as a NumPy array
X = dataset.iloc[:, [3, 4]].values
```

## Using the elbow method to find the optimal number of clusters

```python

from sklearn.cluster import KMeans

# List to store the within-cluster sum of squares (WCSS) for each number of clusters
wcss = []

# Loop over the range of cluster numbers from 1 to 10
for i in range(1, 11):
    # Initialize the KMeans object with the current number of clusters
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    
    # Fit the KMeans algorithm to the data
    kmeans.fit(X)
    
    # Append the WCSS (inertia) to the list
    wcss.append(kmeans.inertia_)

# Plot the WCSS values to visualize the Elbow Method
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## Tarinig the k-means model on the dataset

```python
kmeans =KMeans(n_clusters=5, init='k-means++',)
y_kmeans = kmeans.fit_predict(X)
```

## Visualizing the clusters

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
